```python
#표준 계산기 구현
import tkinter as tk
import math

class Calculator:
    def __init__(self, master):
        self.master = master
        master.title("Calculator")

        # 결과창과 수식창 프레임 생성
        self.result_frame = tk.Frame(master)
        self.result_frame.pack(side=tk.TOP, padx=5, pady=5)
        self.formula_frame = tk.Frame(master)
        self.formula_frame.pack(side=tk.TOP, padx=5, pady=5)

        # 결과창과 수식창 라벨 생성
        self.result_label = tk.Label(self.result_frame, text="0", font=("Arial", 30), anchor=tk.E, width=13)
        self.result_label.pack(side=tk.RIGHT)
        self.formula_label = tk.Label(self.formula_frame, text="", font=("Arial", 20), anchor=tk.E, width=22)
        self.formula_label.pack(side=tk.RIGHT)

        # 버튼 프레임 생성
        self.button_frame = tk.Frame(master)
        self.button_frame.pack(side=tk.TOP, padx=5, pady=5)

        # 버튼 리스트 생성
        self.button_list = [
            ['반올림', '소수점2', '소수점1', '소수점0'],
            ['CUT', 'CE', 'C', '<-'],
            ['n!', 'x^y', 'mod', '/'],
            ['7', '8', '9', '*'],
            ['4', '5', '6', '-'],
            ['1', '2', '3', '+'],
            ['+/-', '0', '.', '=']
        ]

        # 버튼 생성
        self.buttons = {}
        for i, row in enumerate(self.button_list):
            for j, label in enumerate(row):
                if label == '':
                    continue
                button = tk.Button(self.button_frame, text=label, width=5, height=2,
                                   font=("Arial", 15), command=lambda label=label: self.button_click(label))
                button.grid(row=i, column=j)
                self.buttons[label] = button

        self.reset()

    def reset(self):
        self.formula = ""
        self.result = "0"
        self.operator = None
        self.last_button = None
        self.decimal_places = 0
        self.last_number = None

    def button_click(self, label):
        if label in ['+', '-', '*', '/', 'x^y', 'mod']:
        self.operator = label
        self.last_number = float(self.result)
        self.formula += self.result + f" {label} "
        self.result = "0"
        self.decimal_places = 0
    elif label == '+/-':
        if self.result.startswith('-'):
            self.result = self.result[1:]
        else:
            self.result = '-' + self.result
    elif label == '.':
        if self.decimal_places == 0:
            self.result += '.'
            self.decimal_places = 1
    elif label in ['소수점2', '소수점1', '소수점0']:
        self.decimal_places = int(label[-1])
    elif label == '반올림':
        decimal_places = self.decimal_places + 1
        self.result = str(round(float(self.result), decimal_places))
    elif label == 'CUT':
        self.result = str(math.trunc(float(self.result)))
    elif label == 'CE':
        self.result = "0"
    elif label == 'C':
        self.reset()
    elif label == '<-':
        self.result = self.result[:-1] if len(self.result) > 1 else "0"
    elif label == 'n!':
        n = int(float(self.result))
        if n < 0:
            self.result = "Error"
        elif n == 0:
            self.result = "1"
        else:
            self.result = str(math.factorial(n))
    elif label.isdigit():
        if self.result == '0':
            self.result = label
        else:
            self.result += label
    elif label == '=':
        if self.operator is None:
            return
        current_number = float(self.result)
        if self.operator == '+':
            result_number = self.last_number + current_number
        elif self.operator == '-':
            result_number = self.last_number - current_number
        elif self.operator == '*':
            result_number = self.last_number * current_number
        elif self.operator == '/':
            if current_number == 0:
                self.result = "Error"
                return
            result_number = self.last_number / current_number
        elif self.operator == 'x^y':
            result_number = self.last_number ** current_number
        elif self.operator == 'mod':
            result_number = self.last_number % current_number
        self.result = str(result_number)
        self.operator = None
        self.last_number = None
        self.formula = ""
        self.decimal_places = 0
        self.result_label.config(text=self.result)
        self.formula_label.config(text=self.formula)

root = tk.Tk()
calc = Calculator(root)
root.mainloop()



```

```python
#벽돌깨기 프로그램

from tkinter import *
import time
import random

# 게임 화면 생성
window = Tk()
canvas = Canvas(window, width=600, height=400)
canvas.pack()

# 초기 맞힌 블록 수
score = 0

# 맞힌 블록 수를 표시하는 함수
def update_score():
    score_label.config(text=f"Score: {score}")

# 맞힌 블록 수를 증가시키는 함수
def increase_score():
    global score
    score += 1
    update_score()

# 맞힌 블록 수를 표시할 라벨 생성
score_label = Label(window, text="Score: 0")
score_label.pack(side=RIGHT)

# 공 클래스
class Ball():
    def __init__(self, color, size):
        self.id = canvas.create_oval(0, 0, size, size, fill=color)
        self.dx = random.choice([-10, -8, -6, 6, 8, 10])
        self.dy = random.choice([-10, -8, -6, 6, 8, 10])


    # 공 이동 메서드
    def move(self):
        canvas.move(self.id, self.dx, self.dy)
        x0, y0, x1, y1 = canvas.coords(self.id)
        # 공이 화면 밖으로 나가지 않도록 처리
        if y1 > canvas.winfo_height() or y0 < 0:
            self.dy = -self.dy
        if x1 > canvas.winfo_width() or x0 < 0:
            self.dx = -self.dx
            
# 패들 클래스
class Paddle:
    def __init__(self, canvas, x, y):
        self.width = 100 
        self.height = 20
        item = canvas.create_rectangle(x - self.width / 2, y - self.height / 2,
                                       x + self.width / 2, y + self.height / 2,
                                       fill='white')
        self.canvas = canvas
        self.item = item
        self.x = x                      
        self.y = y

    def move(self, dx, dy):
        self.x = self.x + dx
        self.y = self.y + dy
        self.canvas.move(self.item, dx, dy)

    # 패들과 충돌할 때 난반사
    def bounce(self, ball):
        x0, y0, x1, y1 = self.canvas.coords(self.item)
        bx, by = ball.dx, ball.dy
        if x0 < ball.x < x1:
            ball.dy = -by
            ball.dx = bx + random.choice([-3, -2, -1, 0, 1, 2, 3])  # x방향 속도에 난수 추가



# 벽돌 클래스
class Brick():
    def __init__(self, color, width, height):
        self.id = canvas.create_rectangle(0, 0, width, height, fill=color)

    # 벽돌 삭제 메서드
    def delete(self):
        canvas.delete(self.id)
        increase_score()

# 벽돌 생성 함수
def create_bricks():
    bricks = []
    # 6x3 크기의 벽돌 생성
    for row in range(3):
        for col in range(6):
            brick = Brick("red", 90, 30)
            # 벽돌 위치 설정
            x = col * 100 + 10
            y = row * 40 + 10
            canvas.move(brick.id, x, y)
            bricks.append(brick)
    return bricks

# 게임 초기화 함수
def game_init():
    global bricks, paddle
    # 공 3개 생성
    balls = [Ball("blue", 30), Ball("green", 15), Ball("orange", 10)]
    bricks = create_bricks()
    # 패들 생성
    paddle = Paddle(canvas, 300, 350)
    return balls



# 게임 루프 함수
def game_loop(balls):
    global paddle
    # 패들 이동
    if paddle:
        canvas.bind_all('<KeyPress-Left>', lambda event: paddle.move(-10, 0))
        canvas.bind_all('<KeyPress-Right>', lambda event: paddle.move(10, 0))
        
    # 공 이동
    for ball in balls:
        ball.move()
        # 공이 패들과 충돌한 경우
        # if canvas.coords(ball.id)[2] >= paddle.x - paddle.width / 2 and \
        #         canvas.coords(ball.id)[0] <= paddle.x + paddle.width / 2 and \
        #         canvas.coords(ball.id)[3] >= paddle.y - paddle.height / 2 and \
        #         canvas.coords(ball.id)[3] <= paddle.y + paddle.height / 2:
        #     paddle.bounce(ball)
            
        if canvas.coords(ball.id)[2] >= paddle.x - paddle.width / 2 and \
                canvas.coords(ball.id)[0] <= paddle.x + paddle.width / 2 and \
                canvas.coords(ball.id)[3] >= paddle.y - paddle.height / 2 and \
                canvas.coords(ball.id)[1] <= paddle.y + paddle.height / 2:
            paddle.bounce(ball)

            
        # 공이 벽돌에 충돌한 경우
        for brick in bricks:
            if canvas.coords(ball.id)[2] >= canvas.coords(brick.id)[0] and \
                    canvas.coords(ball.id)[0] <= canvas.coords(brick.id)[2] and \
                    canvas.coords(ball.id)[3] >= canvas.coords(brick.id)[1] and \
                    canvas.coords(ball.id)[1] <= canvas.coords(brick.id)[3]:
                # 벽돌 삭제
                brick.delete()
                bricks.remove(brick)
                # 공의 y 방향 반전
                ball.dy = -ball.dy
        # 공이 바닥에 닿았을 경우
        if canvas.coords(ball.id)[3] >= canvas.winfo_height():
            balls.remove(ball)
            canvas.delete(ball.id)
    # 모든 공이 바닥에 닿았을 경우 게임 종료
    if not balls:
        game_over = canvas.create_text(300, 200, text="GAME OVER", fill="red", font=("Helvetica", 40))
    else:
        window.after(50, game_loop, balls)

        
# 게임 초기화
balls = game_init()
# 게임 루프
game_loop(balls)
# 게임 시작
window.mainloop()
```

```python
#2 
import RPi.GPIO as GPIO
import time

Red = 23
Blue = 24
Button = 18
GPIO.setmode(GPIO.BCM)
GPIO.setup(Red, GPIO.OUT)
GPIO.setup(Blue, GPIO.OUT)
GPIO.setup(Button, GPIO.IN, pull_up_down = GPIO.PUD_DOWN)
GPIO.output(Red, False)
GPIO.output(Blue, False)

count = 0
def handler(channel):
    global count
    count += 1
    print("count=", count)
    if count % 2 == 0:
        GPIO.output(Blue, True)
        GPIO.output(Red, False)
        print("BLUE LED ON")
    else:
        GPIO.output(Blue, False)
        GPIO.output(Red, True)
        print("RED LED ON")

GPIO.add_event_detect(Button, GPIO.RISING, callback = handler, bouncetime = 600)

try:
    while True:
        pass
except KeyboardInterrupt:
    print("Program end...\n")
    GPIO.cleanup()

```

```python
#3
import RPi.GPIO as GPIO
import time


Button = 18
Pwm = 23

GPIO.setmode(GPIO.BCM)
GPIO.setup(Pwm, GPIO.OUT)
GPIO.setup(Button, GPIO.IN, pull_up_down = GPIO.PUD_UP)
GPIO.setup(Pwm, GPIO.OUT)
GPIO.output(Red, False)

count = 0
soft_pwm = GPIO.PWM(12, 2)
soft_pwm.start(50)

def handler(channel):
    global count, soft_pwm
    count += 1
    if count > 10:
        count = 0
    dc = (count/10)*100
    print("count = ", count)
    soft_pwm.ChangeDutyCycle(dc)


GPIO.add_event_detect(Button, GPIO.FALLING, callback = handler, bouncetime = 600)

try:
    while True:
        pass
    
except KeyboardInterrupt:
    print("Program end...\n")
    GPIO.cleanup()
```
