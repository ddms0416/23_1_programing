```python
from tkinter import *
from math import *

def click(key):
    if key == '=':
        try:
            calc = entry.get()
            calc = calc.replace("x", "*")
            calc = calc.replace("÷", "/")
            calc_list = []
            num_str = ''
            for c in calc:
                if c.isdigit() or c == '.':
                    num_str += c
                else:
                    if num_str:
                        calc_list.append(float(num_str))
                        num_str = ''
                    calc_list.append(c)
            if num_str:
                calc_list.append(float(num_str))
            
            i = 0
            while i < len(calc_list):
                if calc_list[i] == '*':
                    calc_list[i-1:i+2] = [calc_list[i-1] * calc_list[i+1]]
                elif calc_list[i] == '/':
                    calc_list[i-1:i+2] = [calc_list[i-1] / calc_list[i+1]]
                else:
                    i += 1
            
            result = calc_list[0]
            for i in range(1, len(calc_list), 2):
                if calc_list[i] == '+':
                    result += calc_list[i+1]
                elif calc_list[i] == '-':
                    result -= calc_list[i+1]
                    
            entry.delete(0, END)
            entry.insert(END, str(result))
        except:
            entry.insert(END, "오류")
    elif key == 'C':
        entry.delete(0, END)
    elif key == 'CE': 
        entry.delete(len(entry.get())-1, END) 
    elif key == '+/-': 
        if entry.get()[0] == '-':
            entry.delete(0)
        else:
            entry.insert(0, '-')
    elif key == '√': 
        try:
            result = sqrt(float(entry.get()))
            entry.delete(0, END) 
            entry.insert(END, str(result))
        except:
            entry.insert(END, "오류")
    elif key == 'x²': 
        try:
            result = pow(float(entry.get()), 2)
            entry.delete(0, END) 
            entry.insert(END, str(result))
        except:
            entry.insert(END, "오류")
    elif key == '1/x': 
        try:
            result = 1 / float(entry.get())
            entry.delete(0, END) 
            entry.insert(END, str(result))
        except:
            entry.insert(END, "오류")
    elif key == '%': 
        try:
            result = eval(entry.get())
            entry.delete(0, END) 
            entry.insert(END, str(result/100))
        except:
            entry.insert(END, "오류")
    elif key == 'MC': 
        memory.set('')
    elif key == 'MR': 
        entry.insert(END, memory.get())
    elif key == 'M+': 
        try:
            result = float(entry.get()) + float(memory.get())
            memory.set(str(result))
        except:
            entry.insert(END, "오류")
    elif key == 'M-': 
        try:
            result = float(memory.get()) - float(entry.get())
            memory.set(str(result))
        except:
            entry.insert(END, "오류")
    elif key == '←':
        entry.delete(len(entry.get())-1)
    else:
        entry.insert(END, key)

window = Tk()
window.title("윈도우에서 제공되는 표준계산기 구현")

buttons = [
'MC', 'MR', 'M+', 'M-',
'%', 'CE', 'C', '←',
'1/x', 'x²', '√', '/',
'7',  '8',  '9',  '*',
'4',  '5',  '6',  '-',
'1',  '2',  '3',  '+',
'+/-', '0',  '.', '='
]

memory = StringVar()
memory.set('')
entry = Entry(window, width=33, textvariable=memory, bg="white")
entry.grid(row=0, column=0, columnspan=5, padx=5, pady=5)

for i in range(len(buttons)):
    button = Button(window, text=buttons[i], width=5, height=2, command=lambda key=buttons[i]: click(key))
    row = i // 4 + 1
    col = i % 4
    button.grid(row=row, column=col, padx=5, pady=5)

window.mainloop()
```
